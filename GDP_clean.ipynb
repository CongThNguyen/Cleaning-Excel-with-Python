# Working on Brand Awereness

```python
# Upon close inspection of the data file, there are missing data up to and including row 7th. 
# Since we only concern ourselves with the April month, we can delete all the missing entries.

d2 = pd.read_excel('Brand Awareness.xlsx')
df_brand = d2 


df_brand.head(10)
```

```python
#Drop the first six rows.

df_brand2 = df_brand.drop(labels = range(0, 6), axis=0)

df_brand2
```

```python
# Checking types of object. 
df_brand2.dtypes
```

```python
#Filtered by date and rename.

df_brand3 = df_brand2.query('20170401 <= Month <= 20181001')

df_brand3
```

```python
df_brand3.rename({"Month": "Date"}, axis ="columns", inplace=True)

df_brand4 = df_brand3

# Rename for ease of operations. 
df_brand4

```

# Merge df3 with df_brand4 to create Metric Combined

```python
# Merging

df_merge = pd.merge(df3, df_brand4, how='outer', on='Date')
df_merge.head()
```

# Dealing with Impressions and Clicks

```python
d3 = pd.read_excel('Impression & Clicks.xlsx')
df_clicks = d3 


df_clicks.head(15)
```

```python
df_clicks.dtypes
```

```python
#1. Resampling the data based on weekly frequency. 

#2. Apply it on Date column with 'W' to indicate week. 

#3. and aggregate the Impressions and Clicks week by week.


weekly_data = df_clicks.resample('W', on = 'Date').sum().reset_index().sort_values(by = 'Date')

# weekly_data.head(20)
```

```python
# Change Date column to Week Ending and reassign to a new object. 

weekly_data.rename({"Date": "Week Ending"}, axis ="columns", inplace=True)

Impression_and_Clicks_cleaned = weekly_data

#Show the first 20 rows.
Impression_and_Clicks_cleaned.head(20)
```
